##### Taking care of missing/null values

```python
df['international'] = df['international'].replace(r'-', np.nan)
df['international'] = df['international'].str.replace(',','')
df['international'] = df['international'].astype(float)
```

```python
df['num_students'] = df['num_students'].replace(r'-', np.nan)
df['num_students'] = df['num_students'].str.replace(',','')
df['num_students'] = df['num_students'].astype(float)
```

```python
df['total_score'] = df['total_score'].replace(r'-', np.nan)
df['total_score'] = df['total_score'].astype(float)
```

```python
df['income'] = df['income'].replace(r'-', np.nan)
df['income'] = df['income'].str.replace(',','')
df['income'] = df['income'].astype(float)
```

```python
df.info()
```

```python
freq_snum = df.num_students.dropna().mode()[0]
df['num_students'] = df['num_students'].fillna(freq_snum)
```

```python
avg_ssr = df['student_staff_ratio'].mean()
df['student_staff_ratio'].replace(np.nan, avg_ssr, inplace = True)
```

```python
df['international_students'] = df['international_students'].str.rstrip('%').astype('float') / 100.0
avg_ints = df['international_students'].mean()
df['international_students'].replace(np.nan, avg_ints, inplace = True)
```

```python
df.isnull().sum()
```

```python
df['female_male_ratio'] = df['female_male_ratio'].replace(r'-', np.nan)
if not isinstance(df['female_male_ratio'], float):
    s = df['female_male_ratio'].str.split(':')
    df['female_male_ratio'] = s.str[1].astype(float) / s.str[0].astype(float)
```

```python
avg_fmr = df['female_male_ratio'].mean()
df['female_male_ratio'].replace(np.nan, avg_fmr, inplace = True)
```

```python
avg_ts = df['total_score'].mean()
df['total_score'].replace(np.nan, avg_fmr, inplace = True)
```

```python
freq_inc = df.income.dropna().mode()[0]
df['income'] = df['income'].fillna(freq_inc)
```

```python
freq_int = df.num_students.dropna().mode()[0]
df['international'] = df['international'].fillna(freq_int)
```
