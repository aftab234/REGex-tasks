```python
X = train_data.drop(['id','species'],axis=1).values 
Y = train_data.species
```

```python
X_test = test_data.values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(Y)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
x = sc.transform(X)
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=1)
```

```python
print('Shape of x_train: ',x_train.shape)
print('Shape of x_test: ', x_test.shape)
print('Shape of y_train: ',y_train.shape)
print('Shape of y_test: ', y_test.shape)
```

### Building the model

```python
from sklearn.ensemble import RandomForestClassifier
rdf = RandomForestClassifier(random_state=1)
rdf.fit(x_train,y_train)
```

```python
pred_train = rdf.predict(x_train)
pred_test = rdf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_test,pred_test))
print('Training Accuracy: ', accuracy_score(y_train, pred_train))
print('Testing Accuracy: ', accuracy_score(y_test, pred_test))
```

```python
x_test1 = sc.transform(X_test)
```

```python
y_test1 = rdf.predict_proba(x_test1)
```

```python
submission = pd.DataFrame(y_test1, index=test, columns=encoder.classes_)
```
